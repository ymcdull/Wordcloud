```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba
import io
    
def importStopword(filename = ''):
    stopwords = set()
    with io.open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            stopwords.add(line)
    return stopwords

stopwords = importStopword('stopwords.txt')
# print stopwords

def plot_word_cloud(filename):
    text_from_file_with_apath = open(filename).read()

    wordlist_after_jieba = jieba.cut(text_from_file_with_apath, cut_all = True)
    wl_space_split = " ".join(wordlist_after_jieba)

    my_wordcloud = WordCloud(font_path='simhei.ttf',stopwords=stopwords, collocations=False).generate(wl_space_split)

    plt.imshow(my_wordcloud)
    plt.axis("off")
    plt.show()
```

```python
plot_word_cloud('second_hand_time.txt')
```

```python
plot_word_cloud('strange_thing_in_my_train.txt')
```
